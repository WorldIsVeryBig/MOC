```python
import pandas as pd
 
df = pd.read_excel("./data/excel Data/鹿港龍山寺_後殿Data.xlsx", sheet_name='後殿虎邊束木')
df['資料時間']
```

```python
df['資料時間'] = pd.to_datetime(df['資料時間'])
time_condition = df[df['資料時間'].dt.minute % 10 != 0].index
df.drop(time_condition, inplace=True)
# time_condition_year = df[df['資料時間'] < '2016/12/06 21:30:00' ].index
# df.drop(time_condition_year, inplace=True)    
df['資料時間'] = df['資料時間'].dt.strftime('%Y/%m/%d %H:%M:00')
df = df.reset_index(drop=True)
df = df.drop(columns=['項次', '設備編號', '數值類型'])
df
```

```python
df.to_csv('./data/龍山寺後殿Data/鹿港龍山寺後殿虎邊束木Data.csv', index=False)
```
